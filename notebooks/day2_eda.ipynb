# Benchmark Analysis

## Key Findings

- NIFTY 100 TRI is the most commonly used benchmark.
- Several schemes use NIFTY Midcap and Small Cap indices.
- A few debt-oriented schemes use CRISIL bond indices.
- Benchmarks vary according to the investment objective of each scheme.

## Observation

The dominance of NIFTY-based benchmarks indicates that most schemes track broad Indian equity market performance. Debt schemes use CRISIL bond indices, reflecting fixed-income investment strategies.

```python
pd.crosstab(
    fund_df["fund_house"],
    fund_df["category"]
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cross_tab = pd.crosstab(
    fund_df["fund_house"],
    fund_df["category"]
)

plt.figure(figsize=(10,6))
sns.heatmap(
    cross_tab,
    annot=True,
    cmap="Blues",
    fmt="d"
)

plt.title("Fund House vs Category")
plt.show()
```

# Fund House vs Category Analysis

## Key Findings

- Equity schemes dominate across most fund houses.
- Several fund houses have no debt schemes in the dataset.
- HDFC, ICICI, SBI, Nippon, and Kotak offer both Equity and Debt schemes.
- The distribution reflects the popularity of equity-oriented mutual funds.

## Observation

Most fund houses focus heavily on equity products, while a smaller number maintain a balanced portfolio of both equity and debt funds. This reflects investor preference toward equity-based wealth creation strategies.
